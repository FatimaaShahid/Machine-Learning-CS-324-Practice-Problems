## Implementing Gradient Descent Algorithm

```python
def gradient_descent(derivative_func, learning_rate=0.02,
                     initial_value=0.5, precision=0.001, max_iter=300):
    t=initial_value
    t_list=[]

    for i in range(1,max_iter+1):
        t_list.append(t)
        t = t - learning_rate * derivative_func(t)

        if abs(learning_rate * derivative_func(t))<=precision:
        #if derivative_func(x)==0:
            break

    return t,t_list,i
```

### Partial derivatives for $$ f(t1,t2)=\frac{1}{3^{-t1^2-t2^2}+1} $$ w.r.t. t1 and t2

$$ \frac{\partial f(t1,t2)}{\partial dt1} = \frac{2t1.ln(3).3^{-t1^2-t2^2}}{({3^{-t1^2-t2^2}}+1)^2} $$
$$ \frac{\partial f(t1,t2)}{\partial dt2} = \frac{2t2.ln(3).3^{-t1^2-t2^2}}{({3^{-t1^2-t2^2}}+1)^2} $$

Use online tools for finsing derivatives.
One website is https://www.symbolab.com/solver/derivative-calculator

```python
#Implementing partial derivatives
def pdfx(t1,t2):
    return 2*t1*log(3)*(3**(-t1**2-t2**2))/(3**(-t1**2-t2**2)+1)**2
def pdfy(t1,t2):
    return 2*t2*log(3)*(3**(-t1**2-t2**2))/(3**(-t1**2-t2**2)+1)**2
```

```python
#Custom implementation of gradient descent for two parameters
learning_rate = 0.1
max_iter = 20000
initial_t1 = 1.8
initial_t2=1.5
precision=0.000001

t1_old=initial_t1
t2_old=initial_t2

t1_list=[]
t2_list=[]
for i in range (1,max_iter+1):
    t1_new = t1_old - learning_rate * pdfx(t1_old,t2_old)
    t2_new = t2_old - learning_rate * pdfy(t1_old,t2_old)

    t1_list.append(t1_old)
    t2_list.append(t2_old)

    t1_old=t1_new
    t2_old=t2_new
```

```python
#Plotting the 3D graph
fig=plt.figure(figsize=[16,12])
ax=plt.axes(projection='3d')
ax.set_xlabel('t1', fontsize=16)
ax.set_ylabel('t2', fontsize=16)
ax.set_zlabel('Cost - f(t1,t2)', fontsize=16)
ax.plot_surface(t1,t2,f(t1,t2), alpha=0.4, cmap='summer')
t1_list=np.array(t1_list)
t2_list=np.array(t2_list)
ax.scatter(t1_list,t2_list,f(t1_list,t2_list), alpha=0.4, s=50, color='red')
plt.show()
```

All norm functions are convex

```python
def f(t1,t2):
    return np.sqrt(t1**2 + t2**2)

def pdfx(t1,t2):
    return t1/np.sqrt(t1**2 + t2**2)

def pdfy(t1,t2):
    return t2/np.sqrt(t1**2 + t2**2)
```

part d : run gradient descent

```python
learning_rate = 0.1
max_iter = 5000

t1 = 0.8
t2 = 0.6

for i in range(max_iter):

    grad1 = pdfx(t1,t2)
    grad2 = pdfy(t1,t2)

    t1_new = t1 - learning_rate*grad1
    t2_new = t2 - learning_rate*grad2

    t1 = t1_new
    t2 = t2_new

print("Iterations:", i)
print("t1:", t1)
print("t2:", t2)
print("Cost:", f(t1,t2))
```

At the minimum:the cost function is undefined.

So gradient descent oscillates near (0,0).

```python
eps = 1e-8

def pdfx(t1,t2):
    return t1/(np.sqrt(t1**2 + t2**2)+eps)

def pdfy(t1,t2):
    return t2/(np.sqrt(t1**2 + t2**2)+eps)
#add this small fix to avoid divde by zero error
```

```python
learning_rates = [0.001,0.01,0.05,0.1,0.2,0.5]

costs = []

for lr in learning_rates:

    t1 = 0.8
    t2 = 0.6

    for i in range(5000):

        grad1 = pdfx(t1,t2)
        grad2 = pdfy(t1,t2)

        t1 -= lr*grad1
        t2 -= lr*grad2

    costs.append(f(t1,t2))

plt.plot(learning_rates,costs)
plt.xlabel("Learning Rate")
plt.ylabel("Cost")
plt.show()
```

```python
def f(t1, t2, t3):
    return t1**2 + t2**2 + t3**2

def df_dt1(t1, t2, t3):
    return 2*t1

def df_dt2(t1, t2, t3):
    return 2*t2

def df_dt3(t1, t2, t3):
    return 2*t3
```

```python
def gradient_descent_3d(f, df_dt1, df_dt2, df_dt3,
                        learning_rate=0.02, initial_vals=(0.5, -0.5, 1.0),
                        precision=1e-6, max_iter=5000):
    t1, t2, t3 = initial_vals
    t_list = [(t1, t2, t3)]

    for i in range(1, max_iter+1):
        # Compute partial derivatives
        d1 = df_dt1(t1, t2, t3)
        d2 = df_dt2(t1, t2, t3)
        d3 = df_dt3(t1, t2, t3)

        # Update parameters
        t1 = t1 - learning_rate*d1
        t2 = t2 - learning_rate*d2
        t3 = t3 - learning_rate*d3

        t_list.append((t1, t2, t3))

        # Convergence check
        if abs(d1*learning_rate) < precision and abs(d2*learning_rate) < precision and abs(d3*learning_rate) < precision:
            break

    final_cost = f(t1, t2, t3)
    return (t1, t2, t3), final_cost, i, t_list
```

```python
initial_values = (0.5, -0.5, 1.0)
(final_t1, final_t2, final_t3), final_cost, iterations, t_trace = gradient_descent_3d(
    f, df_dt1, df_dt2, df_dt3, learning_rate=0.02, initial_vals=initial_values
)

print("Converged values:")
print("t1:", final_t1, "t2:", final_t2, "t3:", final_t3)
print("Cost function:", final_cost)
print("Number of iterations:", iterations)
```
